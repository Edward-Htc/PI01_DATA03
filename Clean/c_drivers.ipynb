```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_json('../Datasets/drivers.json')
data.head()
```

```python
data.info()
```

```python
def search_forename(dic):
    return dic['forename']
def search_surmame(dic):
    return dic['surname']
```

```python
forename = data['name'].apply(search_forename)
surname = data['name'].apply(search_surmame)
```

```python
data.insert(loc = 4,column = 'Forename',value = forename)
data.insert(loc = 5,column = 'Surname',value = surname)
```

```python
data.drop(columns='name',axis=1,inplace=True)
```

```python
data.head(15)
```

```python
data['number'].replace('-',np.nan,inplace=True)
```

```python
data['number'].astype('float')
```

```python
data.to_csv('../CleanDataSets/drivers.csv',header=True,index=False)
```
